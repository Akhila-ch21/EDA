```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Importing sample dataset from kaggle - "Nutrition Facts for MC Donald's Menu"

df = pd.read_csv("menu.csv")
print(df.shape)
display(df.head(3))
print(df.columns)

# Getting the basic statistic metrics of data
display(df.describe(include ="all"))

# checking for null values
display(df.isna().sum())

# checking the unique values of category
print(df.Category.nunique(), df.Category.unique())
```

```python
display(df["Sodium"].describe())
# Max sodium value = 3600.00

## Identify the index of the row associated with maximum sodium value (3600.00)
print("\nIndex of row with max sodium value:", df["Sodium"].idxmax())

print(df.at[82,"Item"])
```

## Scatter Plot

### swarmplot - categorywise scatterplot

```python
plot = sns.swarmplot(x = 'Category', y ='Sodium', data = df)

plt.title("Sodium Content")
#plt.xlabel("Category_name"); plt.ylabel("Sodium")

## X-axis lables are rotated at 70 degrees
plt.setp(plot.get_xticklabels(), rotation = 70)

plt.show()
```

### Joint Plot - binary distribution 

- Helps in identifying the "Correlation b/w two variable"

```python
plot = sns.jointplot(x ="Protein", y ="Total Fat", data = df)
plt.show() # Positive Correlation b/w protein & Total Fat
```
